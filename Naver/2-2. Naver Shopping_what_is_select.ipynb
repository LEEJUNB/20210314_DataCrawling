```python
import requests
from bs4 import BeautifulSoup
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
site_url = ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000']
```

# 뽑고자 하는 내용이 동일한 태그에 있으나 url만 다른경우
## 이중 for 반복문 사용하기
- Q) 위처럼 for 문안에 mydata를 선언하면 그 for반복문안에서만 mydata값이 유효한가?

```python
for url in site_url :
    res = requests.get(url,headers=headers)
    soup = BeautifulSoup(res.content,'html.parser')
    mydata = soup.select('#productListArea > ul > li > p > a')
    print(url)
    for item in mydata : 
        print(item.get_text())
```
