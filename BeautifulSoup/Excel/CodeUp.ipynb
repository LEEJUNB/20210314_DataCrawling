```python
import requests 
from bs4 import BeautifulSoup 
import openpyxl
from IPython.display import Image # 이미지 불러오기
```

```python
wb = openpyxl.Workbook() 
sheet = wb.active
sheet.append(["번호", "문제명"])
```

```python
raw = requests.get("https://codeup.kr/problemsetsol.php?psid=10") 
html = BeautifulSoup(raw.text, 'html.parser')
```

```python
# 실패
# container 변수 : 공통적으로 각 번호, 문제명을 포함하는 것.
container = html.select("#problemset > tbody > tr") 
Image("img\CodeUpEach.jpg")
```

```python
# 실패
# container 변수 : 전체 포괄
container = html.select("#problemset > tbody") 
Image("img\CodeUpALL.jpg")
```

```python
for con in container:
    c = con.select_one("div.center").text.strip() # 번호
    l = con.select_one("div.left").text.strip() # 문제명
    
    # sheet 내 각 행에 데이터 추가 
    sheet.append([c, l])
    
wb.save('CodeUp.xlsx')
```
