# CIO Korea 링크에 하이퍼링크 달기

## 보완
- 순위 : enumerate함수를 써서 해결할 것
- 하이퍼링크를 함수로 만들어서 다음 코드에 활용할 것

```python
import requests, openpyxl
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = 'report'
excel_sheet.append(['순서','날짜','제목','링크','내용'])
excel_sheet.column_dimensions['A'].width = 5
excel_sheet.column_dimensions['B'].width = 20
excel_sheet.column_dimensions['C'].width = 80
excel_sheet.column_dimensions['D'].width = 50
excel_sheet.column_dimensions['E'].width = 100

from bs4 import BeautifulSoup
import datetime

dt_now = datetime.datetime.now()

num = 0

res = requests.get('https://www.ciokorea.com/main/')
if res.status_code != 200:
    print('페이지 없음')
else:
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('.news_list_.default_font.color_light_gray.font_12')
    for item in data:
        title = item.select_one('.news_list_title.color_depp_gray.font_limit > a').text
        time = dt_now.date()
        temLink = item.select_one('.news_list_title.color_depp_gray.font_limit > a')["href"]
        link = 'https://www.ciokorea.com'+temLink
        exp = item.select_one('.news_list_summary > a').text
        num = num + 1
        print(num, time, title, link,exp ) # 순위, 날짜, 제목, 링크, 설명 미리 출력해보기
        
        # 엑셀 저장하기
        excel_sheet.append([num, time, title, link, exp]) # 순위, 날짜, 제목, 링크, 설명
        excel_sheet.cell(row=num+1, column=4).hyperlink = link # 하이퍼링크 달기
        
excel_sheet['A1'].alignment = openpyxl.styles.Alignment(horizontal='center') # 순위 셀 가운데 정렬
excel_file.save('[' + str(time) + ']'+' CIO Korea 주요기사.xlsx')
excel_file.close()
```
